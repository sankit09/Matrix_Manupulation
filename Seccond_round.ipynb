```python
df["win_rate"] = df["win_count"]/df["prebid_win_count"]


highest_win_rate = df.sort_values(by='win_rate',ascending=False)

highest_win_rate[['bidder',"prebid_win_count","win_count","win_rate"]].head(1)
```

```python
bidder_stats= df.groupby('bidder').agg({
    'prebid_win_count':'sum',
    'win_count':'sum',
    "request_count":sum
})


bidder_stats['win_rate']= (bidder_stats['prebid_win_count']+bidder_stats['win_count'])/bidder_stats['request_count']
highest_win_rate = bidder_stats.sort_values(by='win_rate',ascending=False)

highest_win_rate.head(2)
```
